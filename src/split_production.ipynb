```python
import pandas as pd
import numpy as np
```

```python
def create_production_period(x):
    weeks = np.zeros(54)
    weeks[x['startA']:x['endA']] = 1
    weeks[x['startB']:x['endB']] = 1
    return pd.Series(weeks.astype(int))
```

```python
index_col = ['Product', 'NUTS_3', 'Region']
df = pd.read_csv('CITRUS18.csv', sep=';', index_col=index_col)
df.head(2)
```

```python
df.shape[0]
```

```python
df['Production_tn'].sum()
```

```python
# Filter out the non-producing
df = df[df['Production_tn'] > 0]
df.head(2)
```

```python
weeks = df.apply(create_production_period, axis=1)
weeks
```

```python
# Divide each row by the total production of that row
# weeks and df are aligned based on their index
num_producing_weeks = weeks.sum(axis=1)

share = df['Production_tn'].div(num_producing_weeks).fillna(0) 

weekly_prod = weeks.mul(share, axis=0)
weekly_prod.columns.name = 'week'
weekly_prod = weekly_prod.stack()
weekly_prod.name = 'Production_tn'
weekly_prod
```
