```python
# https://github.com/tallesfelix/directions-with-osm-networkx/blob/master/Creating%20routes%20with%20networkx%20and%20ipyleaflet.ipynb
import networkx as nx
import osmnx as ox
from shapely.geometry import LineString, mapping
import geopandas as gpd
from ipyleaflet import *
```

```python
place_name = "Attica, Greece"
ox.config(use_cache=True, log_console=True)
#cf = '["highway"~"motorway|motorway_link|trunk|trunk_link|primary|primary_link|secondary|secondary_link|tertiary|tertiary_link"]'
attica_graph = ox.graph_from_place(place_name, network_type='drive') #, custom_filter=cf)
fig, ax = ox.plot_graph(attica_graph)
attica_nodes, attica_edges = ox.graph_to_gdfs(attica_graph)
```

```python
# tests for graph
conn_comps = nx.is_strongly_connected(attica_graph)
weak_comps = nx.is_weakly_connected(attica_graph)
comps = nx.number_attracting_components(attica_graph)
print("is strongly connected: ", conn_comps, " is weakly connected: ", weak_comps, " how many components: ", comps)

# how many nodes?
print(len(attica_nodes))
print(len(attica_edges))
print(nx.number_strongly_connected_components(attica_graph))
#print(max(nx.strongly_connected_components(attica_graph), key=len))
#print(nx.strongly_connected_subgraphs(attica_graph))
#skat = attica_graph.to_undirected()
```
